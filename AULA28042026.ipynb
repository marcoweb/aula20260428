```python
from pymongo import MongoClient
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://brickset.com/sets/year-2023'
```

```python
req = requests.get(url)
html = req.text

print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
items = soup.find_all('article', attrs={'class' : 'set'})
```

```python
for item in items:
    print(item.find('h1').text)
```

```python
kits_info = []
for item in soup.find_all('article', attrs={'class': 'set'}):
    kit = {
        'titulo' : item.find('h1').text,
        'link' : item.find('div', attrs={'class': 'meta'}).find('a')['href']
    }
    kits_info.append(kit)

display(kits_info)
```

```python
client = MongoClient('mongodb://root:rootpass@localhost:27017')
client.server_info()['version']
```

```python
db = client['lego']
collection = db['kits']
```

```python
collection.insert_many(kits_info)
```

```python
for doc in collection.find():
    display(doc)
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('kits.db')
```

```python
cursor = conn.cursor();

cursor.execute("""
    CREATE TABLE IF NOT EXISTS kits(
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        titulo TEXT NOT NULL,
        link TEXT NOT NULL
    )
""")
```

```python
for kit in kits_info:
    cursor.execute(
        "INSERT INTO kits(titulo, link) VALUES (?, ?)",
        (kit['titulo'], kit['link'])
    )

conn.commit()
```

```python
cursor.execute("SELECT * FROM kits")
for item in cursor.fetchall():
    print(item)

conn.close()
```
